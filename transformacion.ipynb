```python
import pandas as pd
```

```python
df1 = pd.read_csv('movies_dataset.csv')
```

```python
print(df1.head(1))
```

```python
df2 = pd.read_csv('credits.csv')
```

```python
#cuando se trata de id es recomendable que sean str
print(df2['id'].dtype)
```

```python
#cambio el tipo de dato para poder unir los df
df2['id'] = df2['id'].astype(str)
```

```python
# union
df = pd.merge(df1, df2, on='id')
```

```python
df.head(1)
```

```python
# desanidar belongs to collection
import ast
df['belongs_to_collection'] = df['belongs_to_collection'].apply(lambda x: ast.literal_eval(x)['name'] if pd.notnull(x) and isinstance(x, str) else 'NaN')
```

```python
df
```

```python
import ast

df['genres'] = df['genres'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else None)
df['genres'] = df['genres'].apply(lambda x: [genre['name'] for genre in x] if isinstance(x, list) else x)

```

```python
df
```

```python
df['spoken_languages'] = df['spoken_languages'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else None)
df['spoken_languages'] = df['spoken_languages'].apply(lambda x: [lang['name'] for lang in x] if isinstance(x, list) else x)
```

```python
import ast

df['production_companies'] = df['production_companies'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])
df['production_companies'] = df['production_companies'].apply(lambda x: [company['name'] for company in x] if isinstance(x, list) else x)

df['production_countries'] = df['production_countries'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])
df['production_countries'] = df['production_countries'].apply(lambda x: [country['name'] for country in x] if isinstance(x, list) else x)

```

```python
df
```

```python
df['cast'] = df['cast'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else None)
df['cast'] = df['cast'].apply(lambda x: [actor['name'] for actor in x] if isinstance(x, list) else x)
```

```python
df
```

```python
df['crew'] = df['crew'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else None)
df['crew'] = df['crew'].apply(lambda x: {crew['job']: crew['name'] for crew in x} if isinstance(x, list) else x)
```

```python
df.head(1)
```

```python
# Esta línea de código convierte el diccionario en una cadena de texto en formato "clave: valor" separada por comas.
# Si el valor no es un diccionario, se mantiene sin cambios.
df['crew'] = df['crew'].apply(lambda x: ', '.join([f"{k}: {v}" for k, v in x.items()]) if isinstance(x, dict) else x)
```

```python
df
```

```python
print(df.dtypes)
```

```python
# punto 2
#Este código se encarga de reemplazar los valores nulos (NaN) en las columnas "revenue" y "budget" 
# del DataFrame df con el valor 0.
df['revenue'] = df['revenue'].fillna(0)
df['budget'] = df['budget'].fillna(0)
```

```python
# punto 3
# elimino las filas del DataFrame df donde la columna "release_date" tiene valores nulos (NaN).
df.dropna(subset=['release_date'], inplace=True)
```

```python
df
```

```python
#punto 4 
# De haber fechas, deberá tener el formato AAAA-mm-dd, además deberá 
# crear la columna release_yeardonde extraerán el año de la fecha de estreno.
```

```python
#Convertir el campo "release_date" al tipo de datos datetime:
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')



```

```python
print(df['release_date'].dtype)
```

```python
# Crea la columna "release_year" extrayendo el año de la fecha de estreno

df['release_year'] = df['release_date'].dt.year
```

```python
df
```

```python
print(df['release_year'].head)
```

```python
#punto 5
#Cree la columna con el retorno de inversión, llame return con los campos revenuey budget, dividiendo estas dos últimas revenue / budget,
# cuando no haya datos disponibles para calcularlo, deberá tomar el valor 0.
```

```python
# identifico el tipo de dato del campo budget
print(df['budget'].dtype)
```

```python
# Convertir el campo "budget" a tipo numérico (flotante)
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
```

```python
# Calcular el ROI
df['return'] = df['revenue'] / df['budget']
```

```python
df
```

```python
# Reemplazar NaN e infinitos en la columna "return" con cero
import numpy as np
df['return'] = df['return'].fillna(0)
df['return'] = df['return'].replace([np.inf, -np.inf], 0)
```

```python
# Redondeamos
df['return'] = df['return'].round(2)
```

```python
df
```

```python
# Punto 6 : Eliminar las columnas que no se utilizarán
#, video, imdb_id, adult, original_title, poster_pathy homepage.
```

```python
# Eliminar las columnas no utilizadas
columns_to_drop = ['video', 'imdb_id', 'adult', 'original_title', 'poster_path', 'homepage']
df.drop(columns=columns_to_drop, inplace=True)

```

```python
df.head()
```

```python
print(df.dtypes)
```

```python
import numpy as np

# Reemplazar [NaN] y [] por NaN en todo el DataFrame
df = df.applymap(lambda x: np.nan if (isinstance(x, list) and len(x) == 0) or x == '[NaN]' else x)
```

```python
df
```

```python
# seguimos mejorando la transformacion de los datos
#df['genres'] = df['genres'].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)

# Convertir listas en las columnas a cadenas
convertir_columnas = ['production_companies', 'production_countries', 'spoken_languages', 'cast', 'genres']

for column in convertir_columnas:
    df[column] = df[column].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)
```

```python
df
```

```python
#verifico si tengo valores duplicados
duplicados = df.duplicated()
duplicados
```

```python
df
```

```python
df.dtypes
```

```python
#algunas correcciones mas 
df['vote_count'] = pd.to_numeric(df['vote_count'], errors='coerce')
df['popularity'] = pd.to_numeric(df['popularity'], errors='coerce')

df['vote_count'] = df['vote_count'].astype(int)
df['popularity'] = df['popularity'].round(2)
```

```python
df.dtypes
```

```python
# Elimino duplicados (filas)
df_nuevo = df.drop_duplicates()
df_nuevo
```

```python
print("Número de filas antes de eliminar duplicados:", len(df))
print("Número de filas después de eliminar duplicados:", len(df_nuevo))
```

```python
#guardo el df 
df_nuevo.to_csv('pelis.csv', index=False)
```
